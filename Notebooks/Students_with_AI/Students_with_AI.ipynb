# AI impact on students

Different relations are made to analyze use case and use intensity by student type.

## 3. Deep analysis

In this section students are grouped by study field to find relations between their study field and performance indicators, also between study field and AI indicators.

### Arts students

A new Dataframe is created to select only students of arts

```python
arts_students = df.loc[df.Major_Category == "Arts"]
arts_students
```

A count of students per year of study is made to know how many there are in each year.

```python
arts_students["Year_of_Study"].value_counts().sort_values()
```

A count of students that have an specific use of AI is made to know what is the most frequent use in each year of study.

```python
arts_s_AI_use = arts_students.groupby(["Year_of_Study","Primary_Use_Case"]).count()
arts_s_AI_use = arts_s_AI_use[["Student_ID"]]
# A sort is made based on Year_of_Study and the count of Student_ID
arts_s_AI_use = (
    arts_s_AI_use
    .reset_index() # Indexes are reset to convert them into columns
    .sort_values(["Year_of_Study", "Student_ID"], ascending=[True, False])
    .reset_index(drop=True)
)
# Filter of columns to avoid showing index columns
arts_s_AI_use = arts_s_AI_use[["Year_of_Study","Primary_Use_Case","Student_ID"]]
arts_s_AI_use
```

This table shows that Copywriting/Drafting is the main use in students of every year category, but the other use cases differ in the different categories. 

With this info it can be given the following statements: 

- Summarizing_Reading is the less important use for Freshman and Sophomore students.
- Debugging/Troubleshooting	represents the 3rd or 4th option among all types of students. 
- Direct_Answer_Generation is the less important use for Graduate and Junior students.
- Ideation seems to be important for Graduate students but not for the other categories.

#### **General stats**

Columns are sorted in a different order to have a clearer view of information, then describe method is called to look for basic statistical indicators.

```python
arts_students_stats = arts_students[['Pre_Semester_GPA', 'Post_Semester_GPA',
                                       'Traditional_Study_Hours', 'Weekly_GenAI_Hours',
                                       'Tool_Diversity', 'Perceived_AI_Dependency',
                                       'Anxiety_Level_During_Exams', 'Skill_Retention_Score']]
arts_students_stats.describe()
```

After analyzing Arts students stats it can be given these preliminary conclusions:

- Traditional study hours remains as the main study method for the students.
- The mean of tool diversity is 2.76, which means that students aren't interested on exploring different types of AI tools.
- Perception of AI dependency by students is low and anxiety level too, so this variables cannot prove a negative impact of AI usage.
- When GPA mean and max on pre and post semester is analyzed it can be concluded that there's no real impact of AI on student performance, that's because GPA post is only 0.2 points higher than GPA pre semester, and the maximum grade remains almost the same.

#### **Burnout risk**

The column Burnout_Risk_Level represents an assessment applied to students to measure risk of having burnout after completing one semester.

A count is made to know that is the most frequent risk type.

```python
arts_burnout_risk = arts_students['Burnout_Risk_Level'].value_counts().reset_index()
arts_burnout_risk['Percentaje'] = round(arts_burnout_risk['count'] / 5933 * 100, 2)
arts_burnout_risk
```

This means that almost half of students have medium risk level of having burnout at the end of semester, and High level only represents 22.68% which is good.

A count of students that have an specific risk level is made to know if there is a variation in the different years of study.

```python
arts_s_year_bnt = arts_students.groupby(["Year_of_Study","Burnout_Risk_Level"]).count()
arts_s_year_bnt = arts_s_year_bnt[["Student_ID"]]
# A sort is made based on Year_of_Study and the count of Student_ID
arts_s_year_bnt = (
    arts_s_year_bnt
    .reset_index() # Indexes are reset to convert them into columns
    .sort_values(["Year_of_Study", "Student_ID"], ascending=[True, False])
    .reset_index(drop=True)
)
# Filter of columns to avoid showing index columns
arts_s_year_bnt = arts_s_year_bnt[["Year_of_Study","Burnout_Risk_Level","Student_ID"]]
arts_s_year_bnt
```

As seen in the table Freshman is the only Year_of_Study category that has a different main risk level.

## Conclusions

After analyzing all statistics it is concluded that there's no real impact of AI usage on college students, this is because there's no significant improve on GPA after finishing one semester, the maximum grade achieved by some students remains the same befor and after finishing one semester. 

The anxiety level and perceived AI dependency is low for all students, the skill retention is on the average, therefore, this leads to conclude that the use of AI tools don't have a negative impact on mental health at least in the short term. 

Impact on mental health cannot be measured with certainty because the mean of traditional study hours is higher than hours spent with AI assistance, so students see AI tools as a complement and another tool for autonomous study.

AI subscription was not considered in this analysis due to lack of information about type of paid plans and amount of money spent on subscriptions, therefore, a relation between paid AI subscription and study improvement cannot be made.

Regarding AI usage, it was found the following: 

- Copywriting/Drafting is the main use case for Arts and Humanities students.
- Ideation is the main use case for Business students.
- Summarizing_Reading is the main use case for Medicine students.
- Debugging/Troubleshooting is the main use case for STEM students.
- Direct_Answer_Generation is not relevant for most of students.
